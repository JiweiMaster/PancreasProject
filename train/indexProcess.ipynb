```python
import numpy as np
import matplotlib.pyplot as plt

ajiMeanList = np.load('999_aji.npy')
ajiMeanListMean = np.mean(ajiMeanList,axis=1)
print('计算得到的平均值: ',ajiMeanListMean)

fig,ax = plt.subplots()
ax.set_xlabel('10 epoch')
ax.set_ylabel('aji value')
ax.set_title('aji index')
ax.plot(ajiMeanListMean)
```
