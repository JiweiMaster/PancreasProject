```python
import numpy as np
import matplotlib.pyplot as plt
import cv2

# 
traindataset = np.load('train/trainDataSet.npy')
labelDataset = np.load('train/labelDataSet.npy')
boundDataset = np.load('train/boundDataSet.npy')
# 
fig,ax = plt.subplots(1,3,figsize=(40,40))
# 
train0 = traindataset[0]
label0 = labelDataset[0]
bound0 = boundDataset[0]
# 
ax[0].imshow(train0)
ax[1].imshow(label0)
ax[2].imshow(bound0)
```
